```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\thoma\Desktop\ALEX PROJECT\Movie Data\movies.csv')
```

```python
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing


for col in df. columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Replace NaNs with a placeholder value (e.g., 0) before converting to integers
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
#Create the correct Year Column
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
df['yearcorrect'] = df['yearcorrect'].fillna(0).astype('int64')
df
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Assuming 'df' is already defined and includes both numeric and non-numeric columns.

# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix for the numeric columns
correlation_matrix = numeric_df.corr(method='pearson')

# Print the correlation matrix
print(correlation_matrix)
```

```python
# Assuming 'df' is your DataFrame containing both numeric and non-numeric columns

# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=[np.number])

# Compute the Kendall correlation matrix for the numeric columns
kendall_corr_matrix = numeric_df.corr(method='kendall')

# Display the Kendall correlation matrix
kendall_corr_matrix
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the Spearman correlation matrix
spearman_corr_matrix = numeric_df.corr(method='spearman')

# Display the correlation matrix
print("\nSpearman Correlation Matrix:")
print(spearman_corr_matrix)
```

```python
# Assuming 'df' is your DataFrame containing both numeric and non-numeric columns

# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix for the numeric columns
correlation_matrix = numeric_df.corr()

# Plot the heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add titles and labels
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

# Display the plot
plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
